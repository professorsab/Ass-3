```python
import joblib
import numpy as np
import requests
from io import BytesIO

# 1. Load model from Hugging Face
model_url = "https://huggingface.co/professorsab/wine-quality-regression-models/resolve/main/ridge_regression_model.pkl"

print("Downloading model from Hugging Face...")
response = requests.get(model_url)
model = joblib.load(BytesIO(response.content))
print("Model downloaded and loaded successfully!\n")

# 2. Define feature names (Wine Quality dataset)
features = [
    "fixed acidity",
    "volatile acidity",
    "citric acid",
    "residual sugar",
    "chlorides",
    "free sulfur dioxide",
    "total sulfur dioxide",
    "density",
    "pH",
    "sulphates",
    "alcohol"
]

# 3. Accept user input for each feature
input_features = []

print("Please enter the following features:")

for feature in features:
    value = float(input(f"Enter {feature}: "))
    input_features.append(value)

# 4. Convert to numpy array and reshape
input_array = np.array(input_features).reshape(1, -1)

# 5. Make prediction
prediction = model.predict(input_array)

print("\nPredicted Wine Quality Score:", prediction[0])
```
